# Working with non-numeric columns

```python
# Count the unique values of all non-numeric columns
non_numeric_cols = df.select_dtypes(exclude=['int64', 'float64']).columns
print(df[non_numeric_cols].nunique())
```

### The following columns should be drop due to high-cardinality (too many unique values)
- Source IP Address
- Destination IP Address
- Payload Data 
- User Information
- Device Information

### These columns will be kept since it could potentially provide useful information
- Timestamp
- Geo-location Data

```python
df = df.drop(columns=["Source IP Address", "Destination IP Address", "Payload Data", "User Information", "Device Information"])
```

```python
# Convert to Timestamp column into Date and Time columns
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Create separate columns
df['Date'] = df['Timestamp'].dt.date
df['Time'] = df['Timestamp'].dt.time
```

```python
# Pop columns and insert at the beginning
date_col = df.pop('Date')
time_col = df.pop('Time')

df.insert(0, 'Date', date_col)
df.insert(1, 'Time', time_col)
```

```python
df = df.drop(columns='Timestamp')
```

```python
df.head()
```

```python
# Check for unique date and time values
print(f"Date: {df['Date'].nunique()}")
print(f"Time: {df['Time'].nunique()}")
```

High-cardinality even after separation => drop both columns

```python
df = df.drop(columns=['Date', 'Time'])
df.head()
```

```python
# Split Geo-location Data column into City and State
df[['City', 'State']] = df['Geo-location Data'].str.split(',', n=1, expand=True)
df['City'] = df['City'].str.strip()
df['State'] = df['State'].str.strip()

df = df.drop(columns='Geo-location Data')
```

```python
# Check for unique City and State values
print(f"City: {df['City'].nunique()}")
print(f"State: {df['State'].nunique()}")
```

- State column has low-cardinality => could be kept to perform encode later on
- City column still possess a high-cardinality => drop

```python
df = df.drop(columns='City')
```

# Working with Object-type columns

```python
df.dtypes
```

```python
df = df.drop(columns='State')
```

```python
target_col = 'Attack Type'

# Find all object-type columns excluding the target column
object_cols = [col for col in df.select_dtypes(include=['object']).columns.tolist()
                if col != target_col]
print(f"Object columns: {object_cols}")
print(f"Total: {len(object_cols)}")

# Check each one
for col in object_cols:
    print(f"\n{col}:")
    print(f"Unique values: {df[col].nunique()}")
    print(f"Sample values: {df[col].unique()[:5]}")
    print(f"Null count: {df[col].isnull().sum()}")
```

```python
df.head()
```

```python
# Separate by number of unique values
binary_cols = []      
low_cardinality = []  

for col in object_cols:
    n_unique = df[col].nunique()
    if n_unique == 2:
        binary_cols.append(col)
    elif n_unique <= 10:
        low_cardinality.append(col)

print(f"\nBinary (2 values): {binary_cols}")
print(f"Low cardinality (3-10): {low_cardinality}")
```

- Columns with 2 unique values => Binary encode

```python
for col in binary_cols:
    print(f"\n{col} values: {df[col].unique()}")
    df[col] = df[col].map({df[col].unique()[0]: 0, df[col].unique()[1]: 1})
```

- Columns with 3-10 unique values => One-Hot Encoding

```python
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

y = df[target_col]
X = df.drop(columns=[target_col])

# Setup the Transformer with dynamic selection
ct = ColumnTransformer(
    transformers=[
        (
            'cat', 
            OneHotEncoder(drop='first', sparse_output=False, dtype=np.int64), 
            make_column_selector(dtype_include='object')
        )
    ], 
    remainder='passthrough'
)

# Set output to pandas for readability
ct.set_output(transform="pandas")

# One-hot encode on object-type features
X_transformed = ct.fit_transform(X)

# Clean up names (remove the 'cat_encoder__' prefix)
X_transformed.columns = [c.split('__')[-1] for c in X_transformed.columns]

# Label encode on target column
le = LabelEncoder()
y_encoded = le.fit_transform(y)

df_final = X_transformed.copy()
df_final[target_col] = y_encoded


print(f"Final dataset shape: {df_final.shape}")
print(f"Columns: {df_final.columns.tolist()}")
print(f"\nTarget column position: {df_final.columns.get_loc(target_col)}")
```
